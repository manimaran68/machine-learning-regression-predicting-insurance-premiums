```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion = "squared_error", n_estimators = 100, max_features = 2)
regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test, y_pred)
```

```python
r_score
```

```python
import pickle
filename = "finalized_model_Random_Forest.sav"
pickle.dump(regressor, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open("finalized_model_Random_Forest.sav", 'rb'))
result = loaded_model.predict([[21, 25.800, 0, 0, 1]])
```
